```python
def generate_codons(seq,frame_no=1):
    #Dividing the sequence into condons(3 nucloetide bases):
    #There are 3 frame shift posibilites
    seq2=[]
    if (frame_no==2):
        for i in range(1,len(seq)-4+1,3):
            seq2.append(seq[i]+seq[i+1]+seq[i+2])
    elif (frame_no==3):
        for i in range(2,len(seq)-4+1,3):
            seq2.append(seq[i]+seq[i+1]+seq[i+2])
    else:
        for i in range(0,len(seq)-3+1,3):
            seq2.append(seq[i]+seq[i+1]+seq[i+2])
    return seq2
    
def find_codon_pos(seq2,start,stop):
    #Finding start codon:
    pos1=[]
    for i in range(len(seq2)):
        if (seq2[i]==start):
            pos1.append(i)
    pos1=pos1[0]*3

    #Finding stop codon:     
    pos2="Not Found"
    pos3="Not Found"
    pos4="Not Found"

    for i in range(len(seq2)):
        if (seq2[i]==stop[0]):
            pos2=i*3
        elif (seq2[i]==stop[1]):
            pos3=i*3
        elif (seq2[i]==stop[2]):
            pos4=i*3
            
    pos=[pos1,pos2,pos3,pos4]
    return pos

def display_loc(pos,start,stop,seq):
    #Displaying Output:  
    print()
    print("\033[1m","Location of codons:","\033[0m")
    print("Total no. of nucleotide bases    : ",len(seq)*3)
    print("Location of start codon    -",start,": ",pos[0])
    print("Location of 1st stop codon -",stop[0],": ",pos[1])
    print("Location of 2nd stop codon -",stop[1],": ",pos[2])
    print("Location of 3rd stop codon -",stop[2],": ",pos[3])
    print()
    
def find_loc(seq,frame_no=1):
    start="ATG"
    stop=["TAA","TAG","TGA"]

    print("\033[1m","Sequence of codons:","\033[0m")
    seq2=generate_codons(seq,frame_no)
    print(seq2)
    pos=find_codon_pos(seq2,start,stop)
    display_loc(pos,start,stop,seq2)
    return seq2,pos
```

#### 2.	How do we know which codon codes for which amino acids ?

<pre>
The universal genetic code is made up of several codons or triplet bases. The standard code has evolved over time to minimize coding errors. The genetic code is degenerate i.e. more than one codon can code for a single amino acid. Due to this, of the 64 codons, 61 codons code for the 20 amino acids.

There are two punctuation marks in the genetic code called the START and STOP codons which signal the end of protein synthesis in all organisms. 

The genetic code can be read in multiple ways depending on where the reading starts. For example, if the base sequence is GGGAAACCC, reading could start from the first letter, G and there will be 3 codons - GGG, AAA, and CCC. If reading starts at G in the second position, the string will have two codons - GGA and AAC. If reading starts at the third base G, 2 codons will again result - GAA and ACC.

Thus, there are 3 ways of reading the code of every strand of genetic material. These different ways of reading a nucleotide sequence is known as a reading frame. Each reading frame will produce a different sequence of amino acids and hence proteins. Thus, in double stranded DNA, there are 6 possible reading frames.

Each codon is a 3-mer which encodes for a specific Amino acid. We can identify which codon codes for which amino acid from the codon chart.
•	In transcription, the DNA sequence of a gene is "rewritten" using RNA nucleotides. In eukaryotes, the RNA must go through additional processing steps to become a messenger RNA, or mRNA.
•	In translation, the sequence of nucleotides in the mRNA is "translated" into a sequence of amino acids in a polypeptide (protein or protein subunit).

As a result, we see tRNA carrying and transferring amino acids to assemble a polypeptide chain.

```python
# creating a new dictionary
dict1 ={"alanine":["GCT", "GCC", "GCA", "GCG"],
            "arginine":["CGT", "CGC", "CGA", "CGG","AGA", "AGG"], 
             "asparagine":["AAT","AAC"],
             "aspartic acid":["GAT", "GAC"],
             "cysteine":["TGT", "TGC"],
             "glutamic acid":["GAA", "GAG"],
             "glutamine":["CAA", "CAG"],
             "glycine":["GGT","GGC", "GGA", "GGG"],
             "histidine":["CAT", "CAC"],
             "isoleucine":["ATT", "ATC", "ATA"],
             "leucine":["TTA", "TTG", "CTT", "CTC","CTA", "CTG"],
             "lysine":["AAA", "AAG"],
             "methionine":["ATG"],
             "phenylalanine":["TTT", "TTC"],
             "proline":["CCT", "CCC", "CCA", "CCG"],
             "serine":["TCT", "TCC", "TCA", "TCG","AGT", "AGC"],
             "threonine":["ACT","ACC", "ACA", "ACG"],
             "tryptophan":["TGG"],
             "tyrosine":["TAT", "TAC"],
             "valine":["GTT", "GTC", "GTA", "GTG"],
            }
```

```python
#Considering only the codons from start codon onwards till first stop codon that appears in the sequence
def seq_between_start_and_stop(seq2,pos):
    #Extracting only the integer elements from pos list
    seq_min=[]
    for i in range(1,len(pos)):
        if(isinstance(pos[i], int)):
            seq_min.append(pos[i])
    
    seq3=[]
    for i in range(int(pos[0]/3),int(min(seq_min)/3)):
        seq3.append(seq2[i]) 
    return seq3

#Search whether a specific codon is in the each amino acid
def search(arr, x): 
    for i in range(len(arr)): 
        if arr[i] == x: 
            return 1 
    return -1

#Gets the amino acid name when a codon is given as input
def get_key(val):
    for key, value in dict1.items():
        if val == value:
            return key
    return "key doesn't exist"

#Finding the amino acid for coresponding codon
def translation_to_amino(seq3,dict1):
    acid_formed="Not found"
    amino_acids=[]
    #List of all 20 amino acids
    key_list = list(dict1.keys())
    
    for codon in seq3:
        for i in key_list:
            list_of_values = dict1[i]
            if (search(list_of_values,codon)==1):
                acid_formed = get_key(list_of_values)
                amino_acids.append(acid_formed)
    return amino_acids

def amino_acids(seq2,pos):
    print("\033[1m","Amino acids produced are:\n","\033[0m")
    seq3 = seq_between_start_and_stop(seq2,pos)
    amino_acids = translation_to_amino(seq3,dict1)
    print(amino_acids)
```
